```python
import requests
import json


def get_chars_after_string(source_str, target_str):
    # 找到目标字符串的索引
    index = source_str.find(target_str)
    if index == -1:  # 如果目标字符串不存在，返回空字符串
        return ""
    # 返回目标字符串之后的所有字符
    return source_str[index + len(target_str):]


def extract_between(s, start, end):
    # 找到开始字符串的索引
    start_index = s.find(start)
    if start_index == -1:  # 如果开始字符串不存在，返回空字符串
        return ""
    # 跳过开始字符串
    start_index += len(start)

    # 找到结束字符串的索引
    end_index = s.find(end, start_index)
    if end_index == -1:  # 如果结束字符串不存在，返回空字符串
        return ""

    # 返回中间的字符串
    return s[start_index:end_index]


# 发送GET请求
def getParticipationCriteriabyNCTid(NCTid):
    response = requests.request('GET', 'https://clinicaltrials.gov/api/v2/studies/' + NCTid + '',
                                headers={'Accept': 'application'})

    # 检查响应状态码
    if response.status_code == 200:
        # 请求成功，打印响应内容
        return json.loads(response.text)
    else:
        # 请求失败，打印错误信息
        print("Error:", response.status_code)


# print(getParticipationCriteriabyNCTid(datalist[0]))
temp = getParticipationCriteriabyNCTid(datalist[0])["protocolSection"]['eligibilityModule']['eligibilityCriteria']
# print(temp)

Inclusion_Criteria = extract_between(temp, 'Inclusion Criteria:', 'Exclusion Criteria:')
print(Inclusion_Criteria)

Exclusion_Criteria = get_chars_after_string(temp, 'Exclusion Criteria:')
print(Exclusion_Criteria)
```

```python
# 输入字符串
def get_condition_items(input_text):
    # 使用分隔符提取术语
    if input_text != 'nan':
        return input_text.split('|')

    # 输出提取的术语


input_text = "Colorectal Carcinoma|Liver Metastases"
print(get_condition_items(input_text))
```

```python
import re

# 输入字符串
input_text = "DIETARY_SUPPLEMENT: probiotic drink with Lactobacillus paracasei K56|DIETARY_SUPPLEMENT: maltodextrin|DIETARY_SUPPLEMENT: Probiotic powder E3|DIETARY_SUPPLEMENT: Probiotic powder E5|DIETARY_SUPPLEMENT: probiotic capsule E7|DIETARY_SUPPLEMENT: probiotic capsule E9|DIETARY_SUPPLEMENT: probiotic capsule E10|DIETARY_SUPPLEMENT: probiotic capsule E11"


def get_Interventions_items(input_text):
# 使用正则表达式提取设备和药物名称
    if input_text:
        Interventions = {"devices": re.findall(r'DEVICE:\s*([^|]+)', input_text),
                         "drugs": re.findall(r'DRUG:\s*([^|]+)', input_text),
                         "produce": re.findall(r'PROCEDURE:\s*([^|]+)', input_text),
                         "behavior": re.findall(r'BEHAVIOR:\s*([^|]+)', input_text),
                         "other": re.findall(r'OTHER:\s*([^|]+)', input_text),
                         "conbinationproduct": re.findall(r'COMBINATION_PRODUCT:\s*([^|]+)', input_text),
                         "dietary_supplement": re.findall(r'DIETARY_SUPPLEMENT:\s*([^|]+)', input_text)}
    return Interventions
# 输出提取的设备和药物名称
print("Interventions:", get_Interventions_items(input_text))
```

```python
clinicalData = []
for index, item in enumerate(datalist):
    # print(  'item' + item[0])
    ParticipationCriteria = getParticipationCriteriabyNCTid(item)["protocolSection"]['eligibilityModule'][
        'eligibilityCriteria']
    Inclusion_Criteria = extract_between(ParticipationCriteria, 'Inclusion Criteria:', 'Exclusion Criteria:')
    Exclusion_Criteria = get_chars_after_string(ParticipationCriteria, 'Exclusion Criteria:')
    if ParticipationCriteria:
        itemClinicalData = {
            "NCTID": item,
            "Study_Title": Study_Title[index],
            "Study_Status": Study_Status[index],
            "Sponsor":Sponsor[index],
            "Collaborators":get_condition_items(str(Collaborators[index])),
            "Conditions": get_condition_items(Conditions[index]),
            "BriefSummary":BriefSummary[index],
            "InclusionCriteria": Inclusion_Criteria,
            "ExclusionCriteria": Exclusion_Criteria,
            "Interventions": get_Interventions_items(str(Interventions[index]))
        }
        clinicalData.append(itemClinicalData)
        print(itemClinicalData)
        print(index)
```

```python
# 保存为Json文件
import json

filename = "图谱2.json"

# 打开文件进行写操作
with open(filename, 'w', encoding='utf8') as file:
    # 使用json.dump()函数将数据转换成JSON字符串后写入文件
    json.dump(clinicalData, file, ensure_ascii=False)

print("已将数据保存至{}".format(filename))
```

```python
# 保存为CSV文件
import csv

headers = ['NCTID', 'Study_Title', 'Study_Status',"Sponsor","Collaborators", 'Conditions', "BriefSummary", 'InclusionCriteria',
           'ExclusionCriteria',"Interventions"]
# 写入CSV文件
with open('fudantupu2.csv', 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=headers)
    writer.writeheader()  # 写入头信息
    for item in clinicalData:
        writer.writerow(item)
```
